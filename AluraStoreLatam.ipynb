### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

pd.read_csv(url)
```

#1. Análisis de facturación



#### En este primer análisis, debes calcular el ingreso total de cada tienda. Sumando los valores de la columna Precio de cada conjunto de datos de la tienda para estimar los ingresos.

```python
filtro1 = tienda["Precio"].sum()
display(filtro1)
```

```python
filtro2 = tienda2["Precio"].sum()
filtro3 = tienda3["Precio"].sum()
filtro4 = tienda4["Precio"].sum()

display(filtro2)
display(filtro3)
display(filtro4)
```

```python
ingresos_por_tienda = pd.Series({
    "Tienda 1": filtro1,
    "Tienda 2": filtro2,
    "Tienda 3": filtro3,
    "Tienda 4": filtro4
})

ingresos_df = ingresos_por_tienda.to_frame(name='Ingresos')

print("Ingresos por Tienda:")
display(ingresos_df.map(lambda x: f"${x:,.0f}"))
```

# 2. Ventas por categoría

#### En este debes calcular la cantidad de productos vendidos por categoría en cada tienda. La idea es agrupar los datos por categoría y contar el número de ventas de cada tipo, mostrando las categorías más populares de cada tienda.

```python
ventas_por_categoria_tienda1 = tienda.groupby('Categoría del Producto').size().sort_values(ascending=False).to_frame(name='Cantidad de ventas').reset_index()

print("Tienda 1:")
display(ventas_por_categoria_tienda1)
```

```python
ventas_por_categoria_tienda2 = tienda2.groupby('Categoría del Producto').size().sort_values(ascending=False).to_frame(name='Cantidad de ventas').reset_index()

print("Tienda 2:")
display(ventas_por_categoria_tienda2)
```

```python
ventas_por_categoria_tienda3 = tienda3.groupby('Categoría del Producto').size().sort_values(ascending=False).to_frame(name='Cantidad de ventas').reset_index()

print("Tienda 3:")
display(ventas_por_categoria_tienda3)
```

```python
ventas_por_categoria_tienda4 = tienda4.groupby('Categoría del Producto').size().sort_values(ascending=False).to_frame(name='Cantidad de ventas').reset_index()

print("Tienda 4:")
display(ventas_por_categoria_tienda4)
```

```python
def obtener_categoria_mas_vendida(df, nombre_tienda):
    categoria_mas_vendida = df.groupby('Categoría del Producto').size().sort_values(ascending=False).head(1)
    nombre_categoria = categoria_mas_vendida.index[0]
    cantidad_vendida = categoria_mas_vendida.values[0]
    return {'Tienda': nombre_tienda, 'Categoría más vendida': nombre_categoria, 'Cantidad de productos vendidos': cantidad_vendida}

top_tienda1 = obtener_categoria_mas_vendida(tienda, 'Tienda 1')
top_tienda2 = obtener_categoria_mas_vendida(tienda2, 'Tienda 2')
top_tienda3 = obtener_categoria_mas_vendida(tienda3, 'Tienda 3')
top_tienda4 = obtener_categoria_mas_vendida(tienda4, 'Tienda 4')

df_categorias_mas_vendidas = pd.DataFrame([top_tienda1, top_tienda2, top_tienda3, top_tienda4])

print("Categoría más vendida por tienda:")
display(df_categorias_mas_vendidas)
```

# 3. Calificación promedio de la tienda


#### En este paso, debes calcular las calificaciones promedio de los clientes para cada tienda. El objetivo es conocer la satisfacción del cliente con los productos vendidos.

```python
calificacion_promedio_tienda1 = tienda['Calificación'].mean()
calificacion_promedio_tienda2 = tienda2['Calificación'].mean()
calificacion_promedio_tienda3 = tienda3['Calificación'].mean()
calificacion_promedio_tienda4 = tienda4['Calificación'].mean()

print(f"Calificación promedio Tienda 1: {calificacion_promedio_tienda1:.2f}")
print(f"Calificación promedio Tienda 2: {calificacion_promedio_tienda2:.2f}")
print(f"Calificación promedio Tienda 3: {calificacion_promedio_tienda3:.2f}")
print(f"Calificación promedio Tienda 4: {calificacion_promedio_tienda4:.2f}")
```

```python
calificaciones_promedio = {
    "Tienda 1": calificacion_promedio_tienda1,
    "Tienda 2": calificacion_promedio_tienda2,
    "Tienda 3": calificacion_promedio_tienda3,
    "Tienda 4": calificacion_promedio_tienda4
}

df_calificaciones_promedio = pd.DataFrame.from_dict(calificaciones_promedio, orient='index', columns=['Calificación Promedio'])

df_calificaciones_promedio_ordenado = df_calificaciones_promedio.sort_values(by='Calificación Promedio', ascending=False)

print("Calificaciones promedio por tienda (ordenado de mejor a peor):")
display(df_calificaciones_promedio_ordenado)
```

# 4. Productos más y menos vendidos

#### En este paso, debes identificar qué productos fueron los más vendidos y los menos vendidos en cada tienda. Visualiza los resultados para que quede claro qué productos destacaron en ventas en cada tienda.

```python
def obtener_producto_extremo(df, nombre_tienda, extremo='mas'):
    conteo_productos = df['Producto'].value_counts()
    if extremo == 'mas':
        producto_extremo = conteo_productos.head(1)
        columna_producto = 'Producto más vendido'
    else:
        producto_extremo = conteo_productos.tail(1)
        columna_producto = 'Producto menos vendido'

    nombre_producto = producto_extremo.index[0]
    cantidad_vendida = producto_extremo.values[0]
    return {'Tienda': nombre_tienda, columna_producto: nombre_producto, 'Cantidad de ventas': cantidad_vendida}
```

```python
productos_mas_vendidos = [
    obtener_producto_extremo(tienda, 'Tienda 1', 'mas'),
    obtener_producto_extremo(tienda2, 'Tienda 2', 'mas'),
    obtener_producto_extremo(tienda3, 'Tienda 3', 'mas'),
    obtener_producto_extremo(tienda4, 'Tienda 4', 'mas')
]

df_productos_mas_vendidos = pd.DataFrame(productos_mas_vendidos)

print("Productos más vendidos por tienda:")
display(df_productos_mas_vendidos)
```

```python
productos_menos_vendidos = [
    obtener_producto_extremo(tienda, 'Tienda 1', 'menos'),
    obtener_producto_extremo(tienda2, 'Tienda 2', 'menos'),
    obtener_producto_extremo(tienda3, 'Tienda 3', 'menos'),
    obtener_producto_extremo(tienda4, 'Tienda 4', 'menos')
]

df_productos_menos_vendidos = pd.DataFrame(productos_menos_vendidos)

print("\nProductos menos vendidos por tienda:")
display(df_productos_menos_vendidos)
```

# 5. Envío promedio por tienda

#### En este paso, debes calcular el costo de envío promedio para cada tienda. El objetivo es comprender cuánto se gasta, en promedio, en el envío de cada tienda.  Sabiendo que el cliente paga el envío.

```python
costo_envio_promedio_tienda1 = tienda['Costo de envío'].mean()
costo_envio_promedio_tienda2 = tienda2['Costo de envío'].mean()
costo_envio_promedio_tienda3 = tienda3['Costo de envío'].mean()
costo_envio_promedio_tienda4 = tienda4['Costo de envío'].mean()

costos_envio_promedio = {
    "Tienda 1": costo_envio_promedio_tienda1,
    "Tienda 2": costo_envio_promedio_tienda2,
    "Tienda 3": costo_envio_promedio_tienda3,
    "Tienda 4": costo_envio_promedio_tienda4
}

df_costos_envio_promedio = pd.DataFrame.from_dict(costos_envio_promedio, orient='index', columns=['Costo de Envío Promedio'])

print("Costo de envío promedio por tienda:")
display(df_costos_envio_promedio.map(lambda x: f"${x:,.2f}"))
```
